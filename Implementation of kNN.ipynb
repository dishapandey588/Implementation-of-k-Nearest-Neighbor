```python
#https://www.datacamp.com/community/tutorials/k-nearest-neighbor-classification-scikit-learn

from math import sqrt
def dist_between(a, b):
    dist_between = 0.0
    for i in range(len(a)-1):
        dist_between += (a[i] - b[i])**2
    return sqrt(dist_between)
 

def neighbor1(train_data, test, knn_neighbors):
    distances = list()
    
    for train_set in train_data:
        dist = dist_between(test, train_set)
        distances.append((train_set, dist))
    distances.sort(key=lambda tup: tup[1])
    neighbor1 = list()
    
    for i in range(knn_neighbors):
        neighbor1.append(distances[i][0])
    return neighbor1
 
    
data = X1
neighbor1 = neighbor1(data, data[0], 4)

for neighbor in neighbor1:
    print(neighbor)
```
